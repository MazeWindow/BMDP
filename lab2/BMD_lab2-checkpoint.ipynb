```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Title")
```

```python
plt.subplot(1,4,1)
plt.plot(x,y,"r")

plt.subplot(1,4,2)
plt.plot(y,x,"g")

plt.subplot(1,4,3)
plt.plot(x,y,"b")

plt.subplot(1,4,4)
plt.plot(x,y,"m")
```

```python
fig=plt.figure()

axes=fig.add_axes([0,0,1,1])

axes.plot(x,y)
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Title")
```

```python
fig=plt.figure()

axes1=fig.add_axes([0,0,1,1])
axes2=fig.add_axes([0.1, 0.5, 0.4, 0.3])

axes1.plot(x,y,"g")
axes1.set_xlabel("x1 label")
axes1.set_ylabel("y1 label")
axes1.set_title("Title1")

axes2.plot(y, x, "m")
axes2.set_xlabel("x2 label")
axes2.set_ylabel("y2 label")
axes2.set_title("Title2")
```

```python
fig, axes=plt.subplots()

axes.plot(x,y, "y")
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Title")
```

```python
fig, axes=plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y,"y")
axes[1].plot(y,x,"m")
```

```python
fig, axes=plt.subplots(nrows=4,ncols=4)
plt.tight_layout()
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.plot(x, x**2, "^g", label="x**2")
ax.plot(x, x**3, "*y", label="x**3")
ax.legend()
```

```python
fig=plt.figure(figsize=(8,4), dpi=100);
```
